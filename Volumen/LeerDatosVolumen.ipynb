```python
import pandas as pd

# Departamento el cual queremos analizar
departamentos = [
    {"name": "Apurimac", "code": "030000"},
    {"name": "Arequipa", "code": "040000"},
    {"name": "Ayacucho", "code": "050000"},
    {"name": "Huancavelica", "code": "090000"},
    {"name": "Huánuco", "code": "100000"},
    {"name": "Ica", "code": "110000"},
    {"name": "Junín", "code": "120000"},
    {"name": "La Libertad", "code": "130000"},
    {"name": "Lima", "code": "150000"},
    {"name": "Pasco", "code": "190000"}
]

# Cargamos los datasets en un diccionario de dataframes para cada departamento antes y después del 2020-05
dfs_pre2005 = {}
for departamento in departamentos:
    file_name = f"{departamento['name']}_volumen_papas_por_mes_pre202005.csv"
    dfs_pre2005[departamento['name']] = pd.read_csv(file_name)

dfs_post2005 = {}
for departamento in departamentos:
    file_name = f"{departamento['name']}_volumen_papas_por_mes_post202005.csv"
    dfs_post2005[departamento['name']] = pd.read_csv(file_name)
    # Concatenar los dataframes correspondientes a cada departamento

dfs_completas = {}
for departamento in departamentos:
    df_pre = dfs_pre2005[departamento['name']]
    df_post = dfs_post2005[departamento['name']]
    dfs_completas[departamento['name']] = pd.concat([df_pre, df_post])

# Guardar los dataframes concatenados en archivos CSV
for departamento in departamentos:
    file_name = f"completas/{departamento['name']}_volumen_papas_por_mes_completo.csv"
    dfs_completas[departamento['name']].to_csv(file_name, index=False)

```

Ahora creamos una sola tabla de las papas a nivel nacional en mega_tabla.csv

```python
mega_tabla = pd.DataFrame(columns=dfs_completas['Apurimac'].columns, index=dfs_completas['Apurimac'].index)
mega_tabla['Fecha'] = dfs_completas['Apurimac']['Fecha']

# Sumar cada celda de cada dataframe por departamento
for departamento in departamentos:
    dfs_completas[departamento['name']] = dfs_completas[departamento['name']].apply(pd.to_numeric, errors='coerce')
    mega_tabla.iloc[:, 1:] = mega_tabla.iloc[:, 1:].fillna(0).add(dfs_completas[departamento['name']].iloc[:, 1:].fillna(0), fill_value=0)


# Guardar la mega tabla en un archivo CSV
mega_tabla.to_csv('completas/mega_tabla_volumen_papas_por_mes.csv', index=False)
```
