```python
import pandas as pd

# Departamento el cual queremos analizar
departamentos = [
    {"name": "Apurimac", "code": "030000"},
    {"name": "Arequipa", "code": "040000"},
    {"name": "Ayacucho", "code": "050000"},
    {"name": "Huancavelica", "code": "090000"},
    {"name": "Huánuco", "code": "100000"},
    {"name": "Ica", "code": "110000"},
    {"name": "Junín", "code": "120000"},
    {"name": "La Libertad", "code": "130000"},
    {"name": "Lima", "code": "150000"},
    {"name": "Pasco", "code": "190000"}
]

# Cargamos los datasets en un diccionario de dataframes para cada departamento antes y después del 2020-05
dfs_pre2005 = {}
for departamento in departamentos:
    file_name = f"{departamento['name']}_volumen_papas_por_mes_pre202005.csv"
    dfs_pre2005[departamento['name']] = pd.read_csv(file_name)

dfs_post2005 = {}
for departamento in departamentos:
    file_name = f"{departamento['name']}_volumen_papas_por_mes_post202005.csv"
    dfs_post2005[departamento['name']] = pd.read_csv(file_name)
```
