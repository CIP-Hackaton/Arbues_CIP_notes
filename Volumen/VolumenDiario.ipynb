# Volumen diario

En este notebook, se proceder√° a extraer datos de volumen diario

```python
import requests
from bs4 import BeautifulSoup
from io import StringIO
import pandas as pd

# URL de la API del MIDAGRI
url = "http://sistemas.midagri.gob.pe/sisap/portal2/mayorista/resumenes/filtrar"

# Headers de la solicitud
headers = {
    "Accept": "*/*",
    "Accept-Encoding": "gzip, deflate",
    "Accept-Language": "en,es;q=0.9",
    "Connection": "keep-alive",
    "Content-Length": "214",
    "Content-Type": "application/x-www-form-urlencoded",
    "Cookie": "autentificador=a3hrlqoj8s5ldfadnqomntcki7",
    "Host": "sistemas.midagri.gob.pe",
    "Origin": "http://sistemas.midagri.gob.pe",
    "Referer": "http://sistemas.midagri.gob.pe/sisap/portal2/mayorista/",
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36",
    "X-Requested-With": "XMLHttpRequest"
}

departamentos = [
    {"name": "Apurimac", "code": "030000"},
    {"name": "Arequipa", "code": "040000"},
    {"name": "Ayacucho", "code": "050000"},
    {"name": "Huancavelica", "code": "090000"},
    {"name": "Hu√°nuco", "code": "100000"},
    {"name": "Ica", "code": "110000"},
    {"name": "Jun√≠n", "code": "120000"},
    {"name": "La Libertad", "code": "130000"},
    {"name": "Lima", "code": "150000"},
    {"name": "Pasco", "code": "190000"}
]
# Inicializamos los DataFrame
precios_papas_por_mes = None
lista_papas_pre202005 = None
lista_papas_post202005 = None

# Iteramos sobre los departamentos
for departamento in departamentos:
    precios_papas_por_mes = pd.DataFrame()
    volumen_papas_por_mes_post202005 = pd.DataFrame()
    volumen_papas_por_mes_pre202005 = pd.DataFrame()

    lista_papas_pre202005 = pd.DataFrame()
    lista_papas_post202005 = pd.DataFrame()

    print(f"Descargando datos de {departamento['name']}")
    for year in range(2015, 2025):
        for month in range(1, 13):
            payload = {
                "mercado": "*",
                "procedencias[]": f"{departamento['code']}",
                "variables[]": "volumen",
                "fecha": f"01/{month:02d}/{year}",
                "desde": "09/12/2015",
                "hasta": "13/12/2018",
                "anios[]": str(year),
                "meses[]": f"{month:02d}",
                "semanas[]": "51",
                # Esta variable indica que es la papa
                "productos[]": "0104",
                "periodicidad": "mensual",
                "_ajax_carga_final": "consulta",
                "ajax": "true"
            }

            # Realizar la solicitud POST
            response = requests.post(url, headers=headers, data=payload)

            # Verificar y mostrar la respuesta
            if response.status_code != 200:
                print(f"Error en la solicitud: {response.status_code}")
                continue


            soup = BeautifulSoup(response.text, "html.parser")

            # Buscar la tabla en el HTML
            table = soup.find("table")  # Encuentra la primera table

            if not table:
                print("No se encontr√≥ ninguna tabla en el HTML.")
                continue
            
        

            # Convertir la tabla HTML en DataFrame
            df = pd.read_html(str(table))[0]
            df = df.loc[:, (df.iloc[2] == 'Total') | (df.iloc[0] == 'Fecha')]

            df.columns = df.iloc[0]
            df = df[3:] 

            # Se separan los datos del data frame por la papa perricholi desde mayo de 2020
            fecha = pd.to_datetime(f"{year}-{month:02d}", format='%Y-%m')
            if fecha < pd.to_datetime('2020-06', format='%Y-%m'):
                volumen_papas_por_mes_pre202005 = pd.concat([volumen_papas_por_mes_pre202005, df], ignore_index=True)
            else:
                volumen_papas_por_mes_post202005 = pd.concat([volumen_papas_por_mes_post202005, df], ignore_index=True)

    # Guardar los datos en un archivo CSV de los valores antes de junio del 2020 y despu√©s
    volumen_papas_por_mes_pre202005.to_csv(f"{departamento['name']}_volumen_papas_por_mes_pre202005.csv", index=False)
    volumen_papas_por_mes_post202005.to_csv(f"{departamento['name']}_volumen_papas_por_mes_post202005.csv", index=False)
```
