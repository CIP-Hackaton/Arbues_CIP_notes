# Nevadas


```python
import pandas as pd

# Lista de nombres de los meses
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
ruta_base = '/content/sample_data/'

# Crear un diccionario para almacenar los DataFrames
dataframes_meses = {}

# Cargar cada archivo en un DataFrame
for mes in meses:
    archivo = f"{ruta_base}{mes}.csv"
    df = pd.read_csv(archivo, delimiter=';')

    # Eliminar columnas sin nombre
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

    # Guardar en el diccionario
    dataframes_meses[mes] = df

# Verificar los DataFrames cargados
for mes, df in dataframes_meses.items():
    print(f"Datos del mes de {mes}:")
    print(df.head(), "\n")
```

```python
# Verificar si un P.E tiene el mismo UBIGEO en cualquiera de las tablas
def verificar_pe_ubigeo(dataframes):
    ubigeo_diferente = []
    for mes1, df1 in dataframes.items():
        for mes2, df2 in dataframes.items():
            if mes1 != mes2:
                merged = pd.merge(df1, df2, on=['NOMBDEP', 'NOMBPROV', 'NOMBDIST'], suffixes=(f'_{mes1}', f'_{mes2}'))
                diferentes = merged[merged['UBIGEO_' + mes1] != merged['UBIGEO_' + mes2]]
                if not diferentes.empty:
                    ubigeo_diferente.append((mes1, mes2, diferentes))
    return ubigeo_diferente

# Frecuencia de valores únicos de cada columna en dos meses cualesquiera
def frecuencia_columnas(df1, df2, mes1, mes2):
    print(f"Frecuencia de valores únicos para el mes de {mes1}:")
    for col in df1.columns:
        print(f"Columna {col}:")
        print(df1[col].value_counts(), "\n")

    print(f"Frecuencia de valores únicos para el mes de {mes2}:")
    for col in df2.columns:
        print(f"Columna {col}:")
        print(df2[col].value_counts(), "\n")

# Información resumen por tabla
def resumen_tablas(dataframes):
    resumen = []
    for mes, df in dataframes.items():
        total_filas = len(df)
        clases_ubigeo = df['UBIGEO'].nunique()
        repeticiones_pe = df.duplicated(subset=['NOMBDEP', 'NOMBPROV', 'NOMBDIST']).sum()

        resumen.append({
            'Mes': mes,
            'Total Filas': total_filas,
            'Clases UBIGEO': clases_ubigeo,
            'Repeticiones P.E.': repeticiones_pe
        })
    return pd.DataFrame(resumen)

# Ejecución de las funciones
# Verificación de UBIGEO diferente entre tablas
resultados_ubigeo = verificar_pe_ubigeo(dataframes_meses)
if resultados_ubigeo:
    for mes1, mes2, diferentes in resultados_ubigeo:
        print(f"Diferencias de UBIGEO entre {mes1} y {mes2}:")
        print(diferentes)

# Frecuencia de valores únicos para dos meses específicos
mes_a, mes_b = 'Enero', 'Febrero'  # Cambia según los meses que quieras comparar
frecuencia_columnas(dataframes_meses[mes_a], dataframes_meses[mes_b], mes_a, mes_b)

# Resumen de cada tabla
resumen = resumen_tablas(dataframes_meses)
resumen
```

```python
# Encontrar la cantidad de UBIGEOs que se repiten en todas las tablas
def ubigeos_repetidos_en_todas(dataframes):
    # Extraer los UBIGEOs de cada DataFrame como conjuntos
    conjuntos_ubigeos = [set(df['UBIGEO']) for df in dataframes.values()]
    # Encontrar la intersección de todos los conjuntos
    ubigeos_comunes = set.intersection(*conjuntos_ubigeos)
    return len(ubigeos_comunes)

# Calcular y mostrar el número de UBIGEOs repetidos en todas las tablas
cantidad_ubigeos_comunes = ubigeos_repetidos_en_todas(dataframes_meses)
print(f"Cantidad de UBIGEOs que se repiten en todas las tablas: {cantidad_ubigeos_comunes}")
```

```python
def resumen_repeticiones(df, mes, columna='UBIGEO'):
    # Calcular la frecuencia de repeticiones de los valores únicos
    conteo_frecuencias = df[columna].value_counts()
    # Contar cuántos valores tienen la misma cantidad de repeticiones
    resumen = conteo_frecuencias.value_counts()

    print(f"Para el mes de {mes}:")
    for repeticiones, cantidad in resumen.items():
        print(f"Hay {cantidad} {columna.lower()}s que se repiten {repeticiones} veces.")

# Ejemplo de uso con el DataFrame del mes de Enero
resumen_repeticiones(dataframes_meses['Enero'], 'Enero', columna='UBIGEO')
```

```python
import re

# Función para extraer rangos de frecuencia de los datos
def obtener_rangos_frecuencia(dataframes):
    rangos = set()
    for df in dataframes.values():
        rangos.update(df['frecuencia'].dropna().unique())  # Ignorar valores nulos

    # Extraer límites numéricos de los rangos usando regex
    rango_valores = {}
    for rango in rangos:
        match = re.match(r'(\d+)\s*-\s*(\d+)', rango)
        if match:
            low, high = map(int, match.groups())
            promedio = (low + high) / 2
            rango_valores[rango] = promedio

    return rango_valores

# Función para consolidar registros duplicados por UBIGEO
def consolidar_frecuencia_por_ubigeo(df, rangos_frecuencia):
    # Mapear la columna 'frecuencia' a valores numéricos basados en los rangos encontrados
    df['frecuencia_valor'] = df['frecuencia'].map(rangos_frecuencia)

    # Consolidar por UBIGEO
    consolidado = (
        df.groupby(['UBIGEO', 'NOMBDEP', 'NOMBPROV', 'NOMBDIST'], as_index=False)
        .agg({'frecuencia_valor': lambda x: x.mean() if not x.isnull().all() else None})  # Manejo de valores nulos
    )

    # Renombrar la columna consolidada como "frecuencia"
    consolidado.rename(columns={'frecuencia_valor': 'frecuencia'}, inplace=True)

    return consolidado

# Obtener los rangos de frecuencia dinámicamente de todas las tablas
rangos_frecuencia = obtener_rangos_frecuencia(dataframes_meses)
print("Rangos de frecuencia detectados:", rangos_frecuencia)

# Aplicar la función a cada tabla de los meses
for mes, df in dataframes_meses.items():
    print(f"Consolidando datos para el mes de {mes}...")
    dataframes_meses[mes] = consolidar_frecuencia_por_ubigeo(df, rangos_frecuencia)

# Verificar los resultados
for mes, df in dataframes_meses.items():
    print(f"Datos consolidados del mes de {mes}:")
    print(df.head(), "\n")
```

```python
def consolidar_todas_las_tablas_con_info(dataframes):
    # Crear una lista para almacenar los DataFrames modificados
    dfs = []

    for mes, df in dataframes.items():
        # Añadir una columna con el nombre del mes y la frecuencia
        df_mes = df[['UBIGEO', 'NOMBDEP', 'NOMBPROV', 'NOMBDIST', 'frecuencia']].copy()
        df_mes.rename(columns={'frecuencia': mes}, inplace=True)
        dfs.append(df_mes)

    # Unir todos los DataFrames por UBIGEO y las demás columnas clave
    df_consolidado = dfs[0]
    for df_mes in dfs[1:]:
        df_consolidado = pd.merge(df_consolidado, df_mes, on=['UBIGEO', 'NOMBDEP', 'NOMBPROV', 'NOMBDIST'], how='outer')

    return df_consolidado

# Consolidar todas las tablas
df_consolidado = consolidar_todas_las_tablas_con_info(dataframes_meses)
```

```python
def rellenar_nulos_con_promedio(df):
    # Obtener las columnas de los meses
    meses = [col for col in df.columns if col not in ['UBIGEO', 'NOMBDEP', 'NOMBPROV', 'NOMBDIST']]

    # Rellenar nulos con el promedio de los meses cercanos
    for i, mes in enumerate(meses):
        for index, row in df.iterrows():
            if pd.isnull(row[mes]):
                # Obtener meses anteriores y posteriores
                valores_vecinos = []
                if i > 0 and pd.notnull(row[meses[i - 1]]):  # Mes anterior
                    valores_vecinos.append(row[meses[i - 1]])
                if i < len(meses) - 1 and pd.notnull(row[meses[i + 1]]):  # Mes posterior
                    valores_vecinos.append(row[meses[i + 1]])

                # Calcular el promedio si hay valores vecinos
                if valores_vecinos:
                    df.at[index, mes] = sum(valores_vecinos) / len(valores_vecinos)

    return df

# Aplicar la función al DataFrame consolidado
df_consolidado_rellenado = rellenar_nulos_con_promedio(df_consolidado)

df_consolidado_rellenado
```
